#### for a good Vocation in Capetown, i am  looking for a low Humidity, no Rain and a average tempareture from 20-25. in the figure above, i have located the three months of april, october and december. 
let us plot april and october and december to decide which one of them has to be our potential vocation month

```python
april = cape.loc[datetime(2014, 4, 1): datetime(2014, 4, 30)]

octo = cape.loc[datetime(2014, 10, 1): datetime(2014, 10, 31)]

dec = cape.loc[datetime(2014, 12, 1): datetime(2014, 12, 31)]
```
